On Shopify, we have exactly 100 sneaker shops, and each of these shops sells only one model of shoe. We want to do some analysis of the average order value (AOV). When we look at orders data over a 30 day window, we naively calculate an AOV of $3145.13. Given that we know these shops are selling sneakers, a relatively affordable item, something seems wrong with our analysis. 

- Think about what could be going wrong with our calculation. Think about a better way to evaluate this data. 

- What metric would you report for this dataset?
- What is its value?


<div class="alert alert-success" role="alert">
    <b> Answer 1: </b> 
    
-There are outliers in the dataset. Therefore, we are getting extreme AOV when it is calculated with Mean since Mean is sensitive to outliers.

- I have used 2 approches to get the average order value. 
    1. Using the Median value instead of Mean value.
    2. Removing all the outliers from the dataset and re-calculate the Mean.
    
    
- The average order value (AOV) calculated in the original dataset using Median: \$284

- The average order value (AOV) calculated with the Mean after all outliers has been removed: \$294
    

Please see below for the detailed calculation.

- When we have outliers, one way to calculate average is using the median since Mean is sensitive to outliers. That gives us \$284 average order value.

**Another way is excluding the outliers from the dataset and re-calculate the average. Let's see what we would get:**

Function to find the outliers in dataset using interquartile range

```python
def find_outliers_IQR(data):

   q1=data.quantile(0.25)

   q3=data.quantile(0.75)

   IQR=q3-q1

   outliers = data[((data<(q1-1.5*IQR)) | (data>(q3+1.5*IQR)))]

   return outliers
```

Outliers in Order_amount

```python
outliers = find_outliers_IQR(data['order_amount'])

print('number of outliers:' + str(len(outliers)))

print('max outlier value:' + str(outliers.max()))

print('min outlier value: '+ str(outliers.min()))
```

Outliers in total_items

```python
outliers = find_outliers_IQR(data['total_items'])

print('number of outliers:' + str(len(outliers)))

print('max outlier value:' + str(outliers.max()))

print('min outlier value: '+ str(outliers.min()))
```

Now, I am excluding the items above 8 and order amount above 735 and recalculate the statistical summary

```python
Data_withoutOutliers= data.query('order_amount < 735' or 'total_items < 8')
```

```python
Data_withoutOutliers.head()
```

```python
Data_withoutOutliers.describe()
```

Here, we got the average order_amount $294 after removing the outliers.

To conclude, for calculating the average order value, we can either use Median value or average value after the outliers has been removed.

a. How many orders were shipped by Speedy Express in total?

<div class="alert alert-success" role="alert">
    <b> Answer a: </b> 
54 orders were shipped by Speedy Express in total.
    
SQL Code below:

```python
%%sql

SELECT 
    ShipperName,
    COUNT(Orders.OrderID) as Total
FROM 
    Orders
    INNER JOIN Shippers on Orders.ShipperID = Shippers.ShipperID
    
WHERE Shippers.ShipperName is "Speedy Express";
```

b. What is the last name of the employee with the most orders?

<div class="alert alert-success" role="alert">
    <b> Answer b: </b> 
"Peacock" is the last name of the employee with the most orders, which total of 40 orders.
    
SQL Code below:

```python
%%sql
SELECT 
	Employees.LastName,
	COUNT(Orders.OrderID)
    
FROM 
	Employees
    INNER JOIN Orders ON Orders.EmployeeID = Employees.EmployeeID
 GROUP BY
 Employees.LastName
```

c. What product was ordered the most by customers in Germany?

<div class="alert alert-success" role="alert">
    <b> Answer c: </b> 
"Steeleye Stout" was ordered the most by customers in Germany (100 quantity).
    
SQL Code below:

```python
%%sql
SELECT 
	OrderDetails.quantity,
    products.ProductName
    FROM Customers
    JOIN Orders ON Orders.CustomerID = Customers.CustomerID
    JOIN OrderDetails ON OrderDetails.OrderID = Orders.OrderID
    JOIN products ON products.ProductID = OrderDetails.ProductID
    

    WHERE
		Customers.Country is "Germany"
GROUP BY products.ProductName
ORDER BY OrderDetails.quantity DESC;
```
